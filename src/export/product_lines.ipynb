```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import os
os.chdir("C:/Users/ibrah/OneDrive/Documents/classicmodel-analysis")
from src.database.connection import get_connection


conn = get_connection()

query = """
    SELECT productLine, 
           SUM(quantityOrdered * priceEach) AS total_sales
    FROM products
    JOIN orderdetails USING(productCode)
    GROUP BY productLine
    ORDER BY total_sales DESC;
    """

df = pd.read_sql(query, conn)

plt.figure()
sns.barplot(data=df, x='total_sales', y='productLine')
plt.title("Ventes par gamme de produits")
plt.tight_layout()
plt.savefig("visualiations/sales_by_product_line.png")
plt.show()

```
