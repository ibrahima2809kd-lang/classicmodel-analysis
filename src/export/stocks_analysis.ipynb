```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import os
os.chdir("C:/Users/ibrah/OneDrive/Documents/classicmodel-analysis")
from src.database.connection import get_connection

conn = get_connection()

query = """
    SELECT productName, quantityInStock
    FROM products
    WHERE quantityInStock < 1000
    ORDER BY quantityInStock ASC;
    """

df = pd.read_sql(query, conn)

plt.figure()
sns.barplot(data=df, x='quantityInStock', y='productName')
plt.title("Produits en stock faible (<1000)")
plt.tight_layout()
plt.savefig("visualisations/low_stock_products.png")
plt.show()
```
