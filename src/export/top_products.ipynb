```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
os.chdir("C:/Users/ibrah/OneDrive/Documents/classicmodel-analysis")

from src.database.connection import get_connection


conn = get_connection()

query = """
    SELECT productName,
           SUM(quantityOrdered) AS total_quantity
    FROM products
    JOIN orderdetails USING(productCode)
    GROUP BY productName
    ORDER BY total_quantity DESC
    LIMIT 10;
    """

df = pd.read_sql(query, conn)

df = df.sort_values(by="total_quantity", ascending=True)

plt.figure()
sns.barplot(data=df, x='total_quantity', y='productName')
plt.title("Top 10 produits les plus vendus")
plt.tight_layout()
plt.savefig("visualisations/top_10_products.png")
plt.show()
```
