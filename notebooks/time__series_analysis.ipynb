```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
os.chdir("C:/Users/ibrah/OneDrive/Documents/classicmodel-analysis")
from datetime import datetime
from src.database.connection import get_connection



conn = get_connection()

query = """
SELECT 
    DATE_FORMAT(orderDate, '%Y-%m') AS month,
    SUM(quantityOrdered * priceEach) AS total_sales
FROM orders
JOIN orderdetails USING(orderNumber)
GROUP BY month
ORDER BY month;
"""

df = pd.read_sql(query, conn)
conn.close()

df['month'] = pd.to_datetime(df['month'])
df['month_number'] = df['month'].dt.month

plt.figure()
sns.boxplot(x='month_number', y='total_sales', data=df)
plt.title("Saisonnalit√© des ventes")
plt.tight_layout()
plt.savefig("visualisations/seasonality.png")
plt.show()
```
